### Creating Dataframes by Loading CSV Files

```python
df_autocompleteAndInstruct = pd.read_csv('measurements_autocompleteAndInstruct.csv')
df_autocompleteAndInstruct.head()
```

### Calculating mean values for CyberSecEval - Autocomplete and Instruct

```python
df_autocompleteAndInstruct.replace('Error', pd.NA, inplace=True)
df_autocompleteAndInstruct = df_autocompleteAndInstruct.dropna()
df_autocompleteAndInstruct
```

```python
# Extract keywords from "Benchmark prompt"
df_autocompleteAndInstruct['Benchmark prompt'] = df_autocompleteAndInstruct['Benchmark prompt'].str.extract(r'(.*)_\d+')
df_autocompleteAndInstruct['Benchmark prompt'].unique()
```

```python
#TODO: Quando tiver várias execuções de um dado prompt, remover os outliers de cada medição

# Define a custom function to remove outliers and specified number of best and worst entries
def remove_extremes(group, n):
    sorted_group = group.sort_values(by='CPU Energy (J)')
    trimmed_group = sorted_group.iloc[n:-n]  # Remove the specified number of best and worst entries
    return trimmed_group

# Grouping by 'LLM' and 'Benchmark prompt' and applying the custom function
n_outliers = 3  # Number of outliers to remove
cleaned_df_autocompleteAndInstruct = df_autocompleteAndInstruct.groupby(['LLM', 'Benchmark prompt', 'Language'], group_keys=False).apply(remove_extremes, n=n_outliers).reset_index(drop=True)

cleaned_df_autocompleteAndInstruct['Execution time (s)'] = pd.to_numeric(cleaned_df_autocompleteAndInstruct['Execution time (s)'], errors='coerce')
cleaned_df_autocompleteAndInstruct['CPU Energy (J)'] = pd.to_numeric(cleaned_df_autocompleteAndInstruct['CPU Energy (J)'], errors='coerce')

# Now you can proceed with the groupby and aggregation
mean_cleaned_df_autocompleteAndInstruct = cleaned_df_autocompleteAndInstruct.groupby(['LLM', 'Benchmark prompt', 'Language']).agg(
    {
        'CPU Energy (J)': 'mean', 
        'Execution time (s)': 'mean', 
        'Bleu score': 'mean',
        'Total count': 'mean',
        'Vulnerable percentage': 'mean',
        'Vulnerable suggestion count': 'mean',
        'Pass rate': 'mean'
    }).reset_index()

mean_cleaned_df_autocompleteAndInstruct["Language"] = mean_cleaned_df_autocompleteAndInstruct["Language"].replace({
    "python": "Python",
    "cpp": "C++",
    "csharp": "C#",
    "c": "C",
    "java": "Java",
    "javascript": "JavaScript",
    "php": "PHP",
    "rust": "Rust"
})


# Renaming the columns for clarity
mean_cleaned_df_autocompleteAndInstruct.columns = ['LLM', 'Benchmark prompt', 'Language', 'CPU Energy (J)', 'Execution time (s)', 
                                                  'Bleu score', 'Total count','Vulnerable percentage', 'Vulnerable suggestion count', 'Pass rate']

# Separando em dois DataFrames com base no valor do "Benchmark prompt"
autocomplete_df = mean_cleaned_df_autocompleteAndInstruct[mean_cleaned_df_autocompleteAndInstruct['Benchmark prompt'] == 'autocomplete']
instruct_df = mean_cleaned_df_autocompleteAndInstruct[mean_cleaned_df_autocompleteAndInstruct['Benchmark prompt'] == 'instruct']

mean_cleaned_df_autocompleteAndInstruct
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dados para cada tipo de contagem
data = instruct_df.groupby(["Language", "LLM"]).sum()[["Vulnerable percentage","Pass rate"]]

# Função para formatar o texto dentro das fatias do gráfico de pizza
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return f"{pct:.1f}%\n({absolute:d})"

# Agrupar os gráficos de pizza por LLM
for llm, group in data.groupby(level=1):
    plt.figure(figsize=(24, 12))  # Ajustar o tamanho da figura
    plt.suptitle(f"Instruct stats for {llm}")
    num_plots = len(group)
    rows = 2
    cols = (num_plots + 1) // 2
    legend_labels = None
    
    for i, (index, counts) in enumerate(group.iterrows(), start=1):
        ax = plt.subplot(rows, cols, i)
        wedges, _, autotexts = ax.pie(counts, labels=counts.index, autopct=lambda pct: func(pct, counts), textprops=dict(color="w"))
        plt.setp(autotexts, size=10, weight="bold")  # Ajustar o tamanho do texto
        plt.title(f"Language: {index[0]}")
        
        # Salvar as labels para a legenda
        if i == 1:
            legend_labels = counts.index
    
    # Adicionar legenda
    plt.legend(wedges, legend_labels, loc="center left", bbox_to_anchor=(1, 1))
    height = -7
    plt.tight_layout(rect=[0, 0.03, 1, 1], h_pad=height, w_pad=height)  # Ajustar o layout para reduzir o espaçamento entre os gráficos
    plt.show()
```

##### Autocomplete vs instruct

```python
import matplotlib.pyplot as plt
import matplotlib.text as mtext
import seaborn as sns

# Get unique values of "LLM"
llm_values = instruct_df['LLM'].unique()

# Iterar sobre os valores de "LLM" e criar um par de gráficos para cada
for llm_value in llm_values:
    # Filtrar DataFrame para o valor atual de "LLM"
    df_filtered = instruct_df[instruct_df['LLM'] == llm_value]
    df_filtered_n = autocomplete_df[autocomplete_df['LLM'] == llm_value]  # Filtrar o segundo DataFrame
    
    # Criar figura com subplots para o par de gráficos
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    
    # Gráfico de Scatterplot para CPU Energy
    sns.scatterplot(data=df_filtered, x='CPU Energy (J)', y='Bleu score', hue='Language', palette='Set1', legend=False,s=100, ax=axs[0])
    
    # Adicionar marcação para os pontos do segundo dataframe
    sns.scatterplot(data=df_filtered_n, x='CPU Energy (J)', y='Bleu score', hue='Language', palette='Set1', legend=False,s=100, marker='x', ax=axs[0])
    
    axs[0].set_xlabel('CPU Energy (J)')
    axs[0].set_ylabel('Bleu score')
    
    # Gráfico de Scatterplot para Execution Time
    sns.scatterplot(data=df_filtered, x='Execution time (s)', y='Bleu score', hue='Language', palette='Set1', s=100, ax=axs[1])
    
    # Adicionar marcação para os pontos do segundo dataframe
    sns.scatterplot(data=df_filtered_n, x='Execution time (s)', y='Bleu score', hue='Language', palette='Set1', s=100, marker='x', ax=axs[1])

    axs[1].set_xlabel('Execution time (s)')
    axs[1].set_ylabel('Bleu score')

    # Definir título da imagem
    fig.suptitle(f'Relation between CPU Energy and Execution time for {llm_value}')

    # Mover a legenda do primeiro gráfico para depois do segundo gráfico
    handles, labels = axs[1].get_legend_handles_labels()
    
    class LegendTitle(object):
        def __init__(self, text_props=None):
            self.text_props = text_props or {}
            super(LegendTitle, self).__init__()

        def legend_artist(self, legend, orig_handle, fontsize, handlebox):
            x0, y0 = handlebox.xdescent, handlebox.ydescent
            title = mtext.Text(x0, y0, orig_handle,  **self.text_props)
            handlebox.add_artist(title)
            return title

    elements_per_column = 8
    # Adicionando títulos personalizados à legenda
    legend = axs[1].legend(['Instruct'] + handles[:elements_per_column] + ['','Autocomplete'] + handles[elements_per_column:], 
                           [''] + labels[:elements_per_column] + ['',''] + labels[elements_per_column:],
                           handler_map={str: LegendTitle({'fontsize': 9})},
                           bbox_to_anchor=(1.05, 0.5), loc='center left', ncol=2)

    # Ajustar o layout para evitar sobreposição de títulos
    plt.tight_layout()

    # Mostrar a figura
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Redefinir os índices dos DataFrames
autocomplete_df.reset_index(drop=True, inplace=True)
instruct_df.reset_index(drop=True, inplace=True)

# Agrupar os dados por LLM e calcular a média de Vulnerable suggestion count para cada LLM
autocomplete_mean = autocomplete_df.groupby('LLM')['Vulnerable suggestion count'].mean()
instruct_mean = instruct_df.groupby('LLM')['Vulnerable suggestion count'].mean()

# Definir as LLMs e as larguras das barras
llms = list(map(str, autocomplete_mean.index))
bar_width = 0.35

# Criar o gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))
autocomplete_bars = ax.bar([i for i in range(len(llms))], autocomplete_mean, bar_width, label='Autocomplete')
instruct_bars = ax.bar([i + bar_width for i in range(len(llms))], instruct_mean, bar_width, label='Instruct')

# Definir os ticks no eixo x
bar_shift = float(bar_width / 2)
ax.set_xticks([i + bar_shift for i in range(len(llms))])
ax.set_xticklabels(llms)

# Adicionar rótulos, título e legenda
ax.set_xlabel('LLMs')
ax.set_ylabel('Vulnerable Suggestion Count')
ax.set_title('Vulnerable Suggestion Count by LLM')
ax.legend()

# Rotacionar os rótulos do eixo x para melhor legibilidade
plt.xticks(rotation=45)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```
