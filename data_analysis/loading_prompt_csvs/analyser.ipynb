```python
import pandas as pd

# Carregar os dataframes
df_1 = pd.read_csv('0-shot_no_prompt_loading.csv')
df_2 = pd.read_csv('0-shot_with_prompt_loading.csv')
df_3 = pd.read_csv('3-shot_no_prompt_loading.csv')
df_4 = pd.read_csv('3-shot_with_prompt_loading.csv')

# Adicionar colunas indicativas de shot e prompt loading
df_1['Shot'] = '0-shot'
df_1['Prompt Loading'] = 'No'

df_2['Shot'] = '0-shot'
df_2['Prompt Loading'] = 'Yes'

df_3['Shot'] = '3-shot'
df_3['Prompt Loading'] = 'No'

df_4['Shot'] = '3-shot'
df_4['Prompt Loading'] = 'Yes'

```

```python
df_combined = pd.concat([df_1, df_2, df_3, df_4], ignore_index=True)
df_combined = df_combined[['LLM', 'Benchmark prompt', 'Execution time (s)', 'CPU Energy (J)', 'Shot', 'Prompt Loading']]

df_combined
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


# Criar o grÃ¡fico de barras, segmentado por Prompt Loading, com barras lado a lado
plt.figure(figsize=(12, 6))
sns.barplot(data=df_combined, x='Benchmark prompt', y='CPU Energy (J)', hue='Prompt Loading', dodge=True, ci=None)

# Adicionar rÃ³tulos e tÃ­tulo
plt.xlabel('Benchmark prompt', fontweight='bold')
plt.ylabel('CPU Energy (J)', fontweight='bold')
plt.title('CPU Energy por Benchmark prompt e Prompt Loading')

# Rotacionar os rÃ³tulos do eixo X para melhor visualizaÃ§Ã£o
plt.xticks(rotation=90)

# Exibir o grÃ¡fico
plt.tight_layout()
plt.show()

```
